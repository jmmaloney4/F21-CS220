```python
import csv

# Modified from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

def cell(csv_header, csv_data, row_idx, col_name):
    col_idx = csv_header.index(col_name)
    val = csv_data[row_idx][col_idx]
    if val == "":
        return None
    
    try: return int(val)
    except ValueError: pass
    
    try: return float(val)
    except ValueError: pass

    return val

def extract_column(csv_header, csv_data, name):
    l = []
    for row in enumerate(csv_data):
         l.append(cell(csv_header, csv_data, row[0], name))
    return l

def unique(l):
    return list(set(l))

def remove_none(l):
    return filter(lambda x: not x is None, l)

def coerce_str(l):
    return list(map(lambda x: str(x), l))

vax_data = process_csv("vaccinations.csv")
vax_header = vax_data[0]
vax_data = vax_data[1:]

vax_header

def vax_cell(row_idx, col_name):
    return cell(vax_header, vax_data)

def vax_col(name):
    return extract_column(vax_header, vax_data, name)
```

```python
#Q3: What are the populations of all the countries in the dataset?
def populations():
    pop = dict(zip(
        vax_col('country'), 
        map(lambda x: None, vax_col('country'))))

    for row in zip(vax_col('country'), vax_col('population')):
        if row[1] != None:
            pop[row[0]] = row[1]
    return pop

populations()
```

```python
def daily_vaccinations_on(given_date):
    l = zip(vax_col('date'), vax_col('country'), vax_col('daily_vaccinations'))
    l = filter(lambda x: x[0] == given_date, l)
    l = dict((map(lambda x: (x[1], x[2]), l)))
    return l
```

```python
 #Q8: How many vaccinations are estimated to have been done across the whole world on each date mentioned in the dataset?

# from functools import reduce

def reduce(f, i, initial = None):
    if initial == None:
        try:
            rv = next(i)
        except StopIteration:
            print("Out")
            return None
    else:
        rv = initial
    
    for j in i:
        rv = f(rv, j)
    return rv

def bucketize_reduce(d, row, key, value):
    k = key(row)
    if not k in d:
        d[k] = []
    d[k].append(value(row))
    return d

def bucketize(l, key = lambda x: x[0], value = lambda x: x[1]):
    return reduce(lambda d, row: bucketize_reduce(d, row, key, value), l, initial = {})


buckets = bucketize(filter(lambda x: x[1] != None, zip(vax_col('date'), vax_col('daily_vaccinations'))))
buckets
dict(map(lambda x: (x[0], sum(x[1])), buckets.items()))
```

```python
# vax_data
NAME_IDX = vax_header.index('country')
DATE_IDX = vax_header.index('date')
POPULATION_IDX = vax_header.index('population')
FULL_VAX_IDX = vax_header.index('people_fully_vaccinated')
POPULATION_DICT = populations()

countries = bucketize(vax_data, key = lambda x: x[NAME_IDX], value = lambda x: x)

def try_int_cov(s):
    try: return int(s)
    except: return s

def country_map(country):
    name = country[0]
    data = country[1]
    # Replace all the '' with None
    data = list(map(lambda row: list(map(lambda x: None if x == '' else x, row)), data))
    
    def most_recent_data(column, date):
        if column == 'population':
            return POPULATION_DICT[name]
        
        if column == 'daily_vaccinations':
            row = list(filter(lambda x: x[DATE_IDX] == date, data))[0]
            return try_int_cov(row[vax_header.index(column)])
        
        d = map(lambda x: (x[DATE_IDX], x[vax_header.index(column)]), data)
        d = list(filter(lambda x: x[0] <= date and x[1] != None, d))
        if len(d) == 0:
            return None
        else:
            max_date = max(d, key = lambda x: x[0])
            return try_int_cov(max_date[1])        
    
    def row_map(row):
        date = row[DATE_IDX]
        row = list(map(lambda x: (x, most_recent_data(x, date)), vax_header))
        return (date, dict(row))
    
    return (name, dict(map(row_map, data.copy())))

countries = dict(map(country_map, countries.items()))
```

```python
#Q17: What is the estimated ratio of people who have been partly vaccinated to people who have been fully vaccinated in each country, as of '10/16/2021'?

ds = map(lambda x: (x[0], x[1]['10/16/2021']), countries.items())

def ratio_map(country):
    name = country[0]
    data = country[1]
    if data['people_vaccinated'] == None or data['people_fully_vaccinated'] == None:
        return (name, None)
    else:
        return (name, (data['people_vaccinated'] - data['people_fully_vaccinated']) / data['people_fully_vaccinated'])

ratio_dict = dict(map(ratio_map, ds))
ratio_dict

```

```python
#Q19: Which country is estimated to have administered the most vaccines per capita, as of '10/14/2021'?
ds101421 = map(lambda x: (x[0], x[1]['10/14/2021']), countries.items())

def ratio_map_percap(country):
    name = country[0]
    data = country[1]
    if data['total_vaccinations'] == None or data['population'] == None:
        return (name, None)
    else:
        return (name, (data['total_vaccinations'] / data['population']))

vpc = filter(lambda x: x[1] != None, map(ratio_map_percap, ds101421))

max(vpc, key = lambda x: x[1])[0]
```
